```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Month': list(range(1, 13)),
    'Seoul': [45, 50, 55, 40, 35, 30, 25, 20, 25, 35, 40, 45],
    'Busan': [40, 45, 50, 35, 30, 25, 20, 15, 20, 30, 35, 40],
    'Daegu': [42, 47, 52, 37, 32, 27, 22, 17, 22, 32, 37, 42],
    'Gwangju': [38, 43, 48, 33, 28, 23, 18, 13, 18, 28, 33, 38],
}

df = pd.DataFrame(data)

df_plot = df.set_index('Month')
df_plot.plot(kind='bar', figsize=(12, 6))

plt.title('Monthly Air Pollution by City')
plt.xlabel('Month')
plt.ylabel('Air Pollution (µg/m³)')
plt.legend(title='City')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
